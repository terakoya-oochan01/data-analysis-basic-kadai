```python
# 課題５

# step1
# numpyrライブラリをインポートする
import numpy as np
# 画像を表示するライブラリをインポートする
import matplotlib.pyplot as pit

# step2
# 描画領域全体に６領域のサブプロットを作成する
fig, axes = pit.subplots(2, 3, tight_layout=True)

# step3
# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0, 1].plot(x1, y1, linewidth=2.0)

axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0, 1].plot()

# ヒストグラム
## 平均が１５、標準偏差５の標準正規分布から2000個の乱数を生成する
x2 = np.random .normal(15, 5, 2000)
pit.xlim(-10, 40)
# ヒストグラムをプロット
axes[0, 2].hist(x2)

# 散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# プロットする
axes[1, 0].scatter(x3, y3)

pit.show()
```
